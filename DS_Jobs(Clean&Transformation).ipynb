```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv("Uncleaned_DS_jobs.csv")
```

```python
df.head(5)
```

```python
df=df.drop(["index"], axis=1)
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.info()
```

```python
df['Salary Estimate'].unique()
```

```python
df['Salary Estimate']=df['Salary Estimate'].str.replace('K (Glassdoor est.)','')
df['Salary Estimate']=df['Salary Estimate'].str.replace('$','')
df['Salary Estimate']=df['Salary Estimate'].str.replace('K','')
```

```python
df['Salary Estimate']
```

```python
df[["Min Salary Estimate", "Max Salary Estimate"]]=df["Salary Estimate"].str.split('-', expand=True)
```

```python
df=df.drop(["Salary Estimate"],axis=1)
```

```python
df['Company Name']
```

```python

df['Company Name'] = df['Company Name'].apply(lambda x: x.split('\n') [0])
```

```python
df['Company Name']
```

```python
 df['Company Name'].value_counts()
```

```python
df['Location']
```

```python
df['State'] = df['Location'].str.split(',', expand=True)[1]
df['State']
```

```python
df=df.drop(['Location'],axis=1)
```

```python
df
```

```python
df['Python']   = df['Job Description'].map(lambda x: 1 if 'python'   in x.lower() else 0)
df['Excel']    = df['Job Description'].map(lambda x: 1 if 'excel'    in x.lower() else 0)
df['Hadoop']   = df['Job Description'].map(lambda x: 1 if 'hadoop'   in x.lower() else 0)
df['Spark']    = df['Job Description'].map(lambda x: 1 if 'spark'    in x.lower() else 0)
df['Aws']      = df['Job Description'].map(lambda x: 1 if 'aws'      in x.lower() else 0)
df['Tableau']  = df['Job Description'].map(lambda x: 1 if 'tableau'  in x.lower() else 0)
df['Big_data'] = df['Job Description'].map(lambda x: 1 if 'big data' in x.lower() else 0)
```

```python
df.head(5)
```

```python
df['Rating'].unique()
```

```python
df['Rating'].dtype
```

```python
df['Rating']=df['Rating'].replace(-1,0)
```

```python
df['Founded'].unique()
```

```python
df['Campany age'] = df['Founded'].apply(lambda x :2024-x if x>0 else 0 )
```

```python
df
```

```python
df['Industry'].value_counts()
```

```python
df[['Industry','Sector']]
```

```python
df['Sector'].value_counts()
```

```python
df['Size'].value_counts()
```

```python
df['Size']=df['Size'].replace('unkown','na')
df['Size']=df['Size'].replace('-1','na')
```

```python
df['Type of ownership'].value_counts()
```

```python
df['Type of ownership']=df['Type of ownership'].replace('-1','na')
```

```python
df["Revenue"].value_counts()
```

```python
df['Revenue']=df['Revenue'].replace('-1','na')
df['Revenue']=df['Revenue'].replace('Unknown / Non-Applicable','na')
```
